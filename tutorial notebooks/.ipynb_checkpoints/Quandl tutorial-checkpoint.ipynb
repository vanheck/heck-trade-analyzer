## Co to je Quandl
Quandl z quandl.com je API, které poskytuje přístup k datám na burze. Quandl obsahuje ticková data akcií, které jsou obsaženy v S&P 500, NASDAQ 100 a Dow Jones 30.

<i>Otázky k vyřešení:</i>
* Nenašel jsem na quandlu intradenní data jednotlivých Indexů, budu to řešit pomocí exportu ze Sierra Chart nebo metatraderu.

### Instalace
V příkazové řádce anaconda spustit příkaz:

pip install quandl

```python
import quandl as q
q.ApiConfig.api_key = open('../etc/quandl-token.key','r').read() # authorize Quandl
```

## Get data

DAX futures (last month) - dayly

```python
dax_data=q.get("CHRIS/EUREX_FDAX1") # get all dayly data of DAX
dax_data=q.get("CHRIS/EUREX_FDAX1", limit=3) # get dayly data of DAX futures
dax_data=q.get("CHRIS/EUREX_FDAX1", start_date='2016-04-20', end_date='2016-04-23') # get dayly data of DAX futures
dax_data
```

## Save data to file

```python
dax_data.to_csv("../data/dax_futures_dayly.csv")
```

## Load data from file

```python
import pandas as pd
pd.read_csv("../data/dax_futures_dayly.csv")
```
